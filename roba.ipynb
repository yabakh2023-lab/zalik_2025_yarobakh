##### 3.створити метод __str__ який буде виводити Ім'я та Групу студента що виконує завдання
##### у циклі створити мінімум 3 об'єкти класу з різними вхідними даними

```python
class StringLengthCalculator:
    def __init__(self, text: str, student_name: str = "Бах Ярослав", group: str = "IT-College"):
        if not isinstance(text, str):
            raise TypeError("текст має бути рядком(str)")
        self.text = text
        self.student_name = student_name
        self.group = group
        self._length = None
    def calculate(self):
        self._length = len(self.text.replace("",""))
    @property
    def length_without_spaces(self):
        return self._length
    def __str__(self):
        return f"Студент: {self.student_name}, Група: {self.group}"
examples = [
    StringLengthCalculator("Hello World"),
    StringLengthCalculator("Python is Great"),
    StringLengthCalculator("Залік з ООП")
]

for obj in examples:
    obj.calculate()
    print(obj)
    print(f"Рядок: '{obj.text}' -> довжина без пробілів: {obj.length_without_spaces}")
    print("_" * 40)
```
